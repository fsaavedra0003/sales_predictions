```python
import pandas as pd
import statsmodels.api as sm


file_path = '/content/data_03.csv'

data = pd.read_csv(file_path)

df = pd.DataFrame(data)

# Set 'Date' column as the index
df.set_index('date', inplace=True)

# Display the original DataFrame
print("Original DataFrame:")
print(df)


# Calculate the percentage change for the 'Value_A' column


#calculate the percentage change between the current and a prior element. .pct_change() This method is particularly useful when dealing with time series data,
#such as stock prices, where you want to analyze the percentage change from one period to the next.
#This is usefull for Time Series Analysis: Understanding the growth or decline rate in time-series data.
#in this case Calculate the percentage change in the 'values' column from the prior value
#the percentage change is calculated as follows: (current value - previous value) / previous value.



df['change'] = df['values'].pct_change()*100

# Reset the index to bring 'Date' back as a column
df.reset_index(inplace=True)

# Display the DataFrame with percentage change for 'values'
print("\nDataFrame with Percentage Change for 'Value_A':")
print(df)






```

```python
#df.reset_index() method in Pandas is used to reset the index of a DataFrame.
#When you perform certain operations or manipulations on a DataFrame, the index might be modified.
df=df.reset_index()
```

```python
df
```

```python
#Create a lag with for loop with 1 additional column. This will shift the values for the particular lag.
for i in range(1,2):
    df['Lag'+str(i)]=df['change'].shift(i)
```

```python
df
```

```python
#Removing the NaN values.
df=df.dropna()
```

```python
df
```

```python
#Add the direction columns (check if values is positive or negative in "Today" column )
#This is done by adding a new column called Direction and adding the respective result 1 or 0
df['direction']=[1 if i > 0 else 0 for i in df['change']]
```

```python
df
```

```python
#Adding a constant term allows the linear regression model to have an intercept term.
#Including the constant term ensures that the model accounts for the intercept in its predictions.
df = sm.add_constant(df)
```

```python
df
```

```python
X=df[['const', 'Lag1']]
```

```python
y= df.direction
```

```python
model = sm.Logit(y, X)
```

```python
result = model.fit()
```

```python
result.summary()
```

```python
prediction = result.predict(X)
```

```python
prediction
```

```python
def cofusion_matrix(act,pred):
    predtrans=["Up"if i > 0.5 else "Down" for i in pred]
    actuals=["Up"if i > 0 else "Down" for i in act]
    cofusion_matrix = pd.crosstab(pd.Series(actuals),
                                  pd.Series(predtrans),
                                  rownames=['Actual'],
                                  colnames=['Predicted'])
    return cofusion_matrix
```

```python
cofusion_matrix(y,prediction)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix


# Create and fit the logistic regression model
logreg_model = LogisticRegression()
logreg_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = logreg_model.predict(X_test)

# Evaluate the model's performance
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

# Print the evaluation metrics
print(f"Accuracy: {accuracy}")
print(f"Confusion Matrix:\n{conf_matrix}")
print(f"Classification Report:\n{classification_rep}")
y_pred
```
